# Handling with Missing Values

Before doing anything, Saving a checkpoint at first :v

```python
nonmv_df = cleaned_df.copy()
nonmv_df.head()
```

Firstly, we need to demonstrate how impact of the missing values on our dataset.

```python
# Percentage of missing values of each columns
percent_mv_df = pd.DataFrame(columns=('Column', 'Missing value'))
for col in cleaned_df.columns:
    percent_mv_df = percent_mv_df.append({'Column':col, 'Missing value':cleaned_df[col].isnull().sum()/cleaned_df.shape[0] * 100}, 
                          ignore_index=True)
percent_mv_df.sort_values(by='Missing value', ascending=False, inplace=True)
percent_mv_df.drop(labels=percent_mv_df[percent_mv_df['Missing value']==0].index, axis=0, inplace=True)
percent_mv_df.reset_index(inplace=True, drop=True)
percent_mv_df
```

First of all, we would drop the **big 4 + 1** [`PoolQC`, `MiscFeature`, `Alley`, `Fence` and `LotFrontage`]. But, the point here is: Will we miss any important data? According to this [notebook that is written by 
Pedro Marcelino](https://www.kaggle.com/pmarcelino/comprehensive-data-exploration-with-python/notebook), none of these variables seem to be very important, since most of them are not aspects in which we think about when buying a house (maybe that's the reason why data is missing?). Then, we will happy to delete them.

Next, we take a look at `Garage` brothers. According to [Pedro Marcelino](https://www.kaggle.com/pmarcelino/comprehensive-data-exploration-with-python#4.-Missing-data), since the most important information regarding **garages** is expressed by `GarageCars`, I'll delete the mentioned `GarageX` variables. The same logic applies to `BsmtX` variables.

In addition, `MasVnrType` and `MasVnrArea` seem trivial and they have a quite strong correlation with `OverallQual` which is already considered. Hence, dropping them do not loose any information.

The last one `Electrical`, we just delete some observations that contain missing values instead of dropping the whole column.

In conclusion, we will dropping all variables, except `Electrical`.

```python
nonmv_df.drop(labels=nonmv_df[percent_mv_df[percent_mv_df['Missing value'] > 0.1].Column], axis=1, inplace=True)
nonmv_df.drop(labels=nonmv_df[nonmv_df['Electrical'].isnull()].index, axis=0, inplace=True)
```

# Outliers
To identify the outliers, I will use the bivariate analysis method. Namely, we use scatter plot to discover something that is STRANGE.

```python
num_df = nonmv_df.select_dtypes(include=['int64', 'float64']).copy()
num_df = pd.melt(num_df, id_vars=['SalePrice'])
num_df.head()
```

```python
def FacetGrid_RegPlot(x, y, **kwargs):
    sns.regplot(x=x, y=y, ci=95, scatter_kws={'alpha':0.3, 's':50})
```

```python
g = sns.FacetGrid(num_df, col='variable', col_wrap=3, sharex=False, sharey=False, height=5)
g.map(FacetGrid_RegPlot, 'value', 'SalePrice')
```

Looking closer to the plots, we easily notice some strange points that do not follow the trend and stand alone. Yes, that's it! In `GrLivArea`/`SalePrice` plot, there are 2 points with high `GrLivArea` that are eligible for tossing away, and they also seem to be the outliers in others plot such as: `BsmtFinSF1`, `TotalBsmtSF`, etc.

```python
nonmv_df.drop(labels=nonmv_df[nonmv_df.loc[:, 'GrLivArea'] > 4500].index, inplace=True, axis=0)
nonmv_df[nonmv_df.loc[:, 'GrLivArea'] > 4500]
```

Let's plot again to make sure that we already deleted them!

```python
num_df = nonmv_df.select_dtypes(include=['int64', 'float64']).copy()
num_df = pd.melt(num_df, id_vars=['SalePrice'])
g = sns.FacetGrid(num_df, col='variable', col_wrap=3, sharex=False, sharey=False, height=5)
g.map(FacetGrid_RegPlot, 'value', 'SalePrice')
```

**LOG TRANSFORMATION**

```python
for col in skewed_col:
   nonoutliers_df.loc[nonoutliers_df[col] > 0, col] = power_transform(np.reshape(nonoutliers_df.loc[nonoutliers_df[col] > 0, col].tolist(), (-1,1)), method='box-cox')
```

```python
nonoutliers_df[skewed_col].hist(figsize=(20, 20), grid=False)
```

```python
nonskewed_df = nonoutliers_df.copy()
nonskewed_df.info()
```

# Encode Categorical values

```python
encoded_df = nonskewed_df.copy()
```

```python
# Numerical columns
encoded_df.select_dtypes(include=('int64', 'float64')).columns
```

```python
# Categorical columns
cat_cols = encoded_df.select_dtypes(exclude=('int64', 'float64')).columns
cat_cols
```

### Ordered Categorical

```python
ordered_categorical_cols = ['ExterQual', 'ExterCond', 'HeatingQC', 'KitchenQual']
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(['None', 'Po', 'Fa', 'TA', 'Gd', 'Ex'])
```

```python
for col in ordered_categorical_cols:
    encoded_df[col] = label_encoder.transform(encoded_df[col])
encoded_df[ordered_categorical_cols]
```

### Unordered Categorical

```python
unordered_categorical_cols = cat_cols.drop(labels=ordered_categorical_cols)
unordered_categorical_cols
```

```python
onehot_encoder = OneHotEncoder(drop='first', handle_unknown='error', sparse=False)
onehot_features = onehot_encoder.fit_transform(encoded_df[unordered_categorical_cols])
```

```python
# Make a one-hot dataframe 
onehot_df = pd.DataFrame(data=onehot_features, columns=onehot_encoder.get_feature_names(input_features=unordered_categorical_cols))
```

```python
onehot_df.reset_index(drop=True, inplace=True)
encoded_df.reset_index(drop=True, inplace=True)
```

```python
# Remove the original columns and replace them with one hot features 
encoded_df.drop(labels=unordered_categorical_cols, axis=1, inplace=True)
encoded_df = pd.concat(objs=[encoded_df, onehot_df], axis=1, ignore_index=False)
encoded_df.tail()
```

```python
encoded_df.info()
```
